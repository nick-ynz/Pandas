### 对于未在read_csv处设定日期的处理方法

```python
import pandas as pd
import numpy as np

pd.set_option('expand_frame_repr', False)

df = pd.read_csv(filepath_or_buffer='stock_data.csv',
                encoding='gbk',
                # parse_dates=['交易日期']
                )

df['交易日期'] = pd.to_datetime(df['交易日期'])

print(df.at[0,'交易日期'])
print(type(df.at[0,'交易日期']))
```

```python
df = df[['交易日期', '股票代码', '开盘价', '最高价', '最低价', '收盘价', '成交量', '成交额']]
df = df[df['交易日期'] > pd.to_datetime('20171201')]
df = df[df['交易日期'] < pd.to_datetime('20190101')]
df['前收盘价'] = df.groupby('股票代码')['收盘价'].shift(1) 

df.drop(df[np.isnan(df['前收盘价'])].index, inplace=True) # 定位NaN所在的index后删除行
print(df)
```

```python
print(df['交易日期'].dt.day.head()) # year, month, hour, minute, second
```

```python
print(df['交易日期'].dt.dayofyear.head()) 
# 一年中第几天 dayofweek, weekday, weekday_name
```

```python
print(df['交易日期'].dt.days_in_month.head()) # 所在月份有几天
```

```python
print(df['交易日期'].dt.is_month_end.head()) # 是否为月末 is_month_start
```
