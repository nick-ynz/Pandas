# pandas库

### pandas导入数据

```python
import pandas as pd

df = pd.read_csv(
    filepath_or_buffer='sh000001.csv', # (此为相对路径，对同一目录下所处的文件有效)请修改为自己的csv文件路径，其中Windows系统下为转义字符,在左单引号前添加r才能正常导入，如 r'C:\\……' ，否则需将每个\改为\\
    encoding='gbk',
    sep=',',
    #skiprows=1,                  #跳过行数
    #nrows=5,                     #只读取除表头外前5行数据
    parse_dates=['交易日期'],      #识别为日期
    #index_col=['交易日期'],       #设置index
    #usecols=['交易日期','开盘价'], #只读入某些列
    #error_bad_lines=False,        #数据脏乱时不报错
    #na_values='NULL'             #当数据有NULL时，当做空值
)
print(df)
```

### 查看数据函数

```python
print(df.shape)
print(df.shape[0])
print(df.shape[1])
print(df.index)
print(df.columns)
print(df.dtypes)
print(df.head(3))
print(df.tail(3))
print(df.sample(n=3))#参数改成frac=0.3可取比例
print(df.describe())
```

```python
pd.set_option('expand_frame_repr', False) #使表格在呈现所有列（否则列数过多会出现类似以下行的省略）
print(df)
```

```python
pd.set_option('display.max_rows', 100) #显示100行
print(df)
```

```python
pd.set_option('precision', 1) #浮点数显示精度
print(df)
```
