### 以交易日期为index导入数据
### 不将交易日期识别为日期（字符串类型）

```python
df = pd.read_csv(
    filepath_or_buffer='sh000001.csv',
    encoding='gbk',
    sep=',',
    #skiprows=1,                  #跳过行数
    #nrows=5,                     #只读取除表头外前5行数据
    #parse_dates=['交易日期'],      #识别为日期
    index_col=['交易日期'],       #设置index
    #usecols=['交易日期','开盘价'], #只读入某些列
    #error_bad_lines=False,        #数据脏乱时不报错
    #na_values='NULL'             #当数据有NULL时，当做空值
)
print(df)
```

### loc，at函数 以表的行列名称查看数据

```python
print(df.loc['1991-11-29'])
```

```python
print(df.loc[['1991-11-29', '1992-02-28']])
```

```python
print(df.loc['1991-11-29':'1992-02-28'])
```

```python
print(df.loc[:,'开盘价':'收盘价'])
```

```python
print(df.at['2000-03-31','开盘价'])
```

### iloc函数 以表的position查看数据

```python
#iloc 通过position查看数据
print(df.iloc[0,3])
```

```python
print(df.iloc[0])
```

```python
# loc包含冒号后者，iloc不包括
print(df.iloc[1:3])
```

```python
print(df.iloc[:,0:2])
```
