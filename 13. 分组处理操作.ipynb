### groupby函数

```python
import pandas as pd

df = pd.read_csv('stock_data.csv', encoding='gbk', parse_dates=['交易日期'])

print(df)
```

```python
print(df.groupby('交易日期').size())
print('-' * 50)
print(df.groupby('股票代码').size())
```

```python
print(df.groupby('交易日期').get_group('20010430'))
print('-' * 50)
print(df.groupby('股票代码').get_group('sh600000'))
```

```python
# print(df.groupby('股票代码').describe())
# print(df.groupby('股票代码').head(3))
# print(df.groupby('股票代码').tail(3))
# print(df.groupby('股票代码').first())
# print(df.groupby('股票代码').last())
print(df.groupby('股票代码', as_index=False).nth([2, 5])) # 任取行数 as_index不让股票代码作为index
```

```python
# print(df.groupby('股票代码')['收盘价', '成交量'].mean())
# print(df.groupby('股票代码')['收盘价', '成交量'].max())
# print(df.groupby('股票代码')['收盘价', '成交量'].sum())

print(df.groupby('交易日期')['成交量'].rank())
```

```python
df.loc[df['交易日期'].dt.month < 7, '年份'] = '上半年'
df['年份'].fillna(value='下半年', inplace=True)
print(df)
```

```python
print(df.groupby(['股票代码', '年份']).size())
```

```python
# 也可使用
for code, group in df.groupby('股票代码'):
    print(code)
    print(group[['交易日期', '股票代码']].head(3))
    if(code == 'sh600010'):
        break
```
