## 批量导入数据

### 遍历文件夹方法

```python
import pandas as pd
import os

pd.set_option('expand_frame_repr', False)

file_location = '.' # 当前文件夹

for root, dirs, files in os.walk(file_location):
    print('当前文件夹：', root)
    print('包含的文件夹：', dirs)
    print('包含的文件：', files)
```

### 遍历sample文件下，获取文件名后缀为.csv的文件，用os.path方法获得文件的路径，用append获得文件名list

```python
file_list = []

file_location = './sample' # sample文件夹


for root, dirs ,files in os.walk(file_location):
    for filename in files:
        if filename.endswith('.csv'):
            file_path = os.path.join(root, filename)
            file_path = os.path.abspath(file_path)
            file_list.append(file_path)
print(file_list)
```

### 以文件名寻找相关csv文件，填充进all_data中

```python
all_data = pd.DataFrame()
for fp in sorted(file_list):
    print(fp)
    df = pd.read_csv(fp, encoding='gbk')
    all_data = all_data.append(df)
    all_data = all_data.append(df, ignore_index=True)
    
print(all_data)
```

### 排序

```python
all_data.sort_values(by=['交易日期', '股票代码'], inplace=True)
print(all_data)
```

### 输出

```python
all_data.to_csv('all_data.csv', index=False, encoding='gbk')
```

```python
# hdf文件
all_data.to_hdf(
    'a_stock.h5',
    key='all_data',
    mode='w'
    )
```
