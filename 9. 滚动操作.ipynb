### 导入上证指数数据

```python
import pandas as pd

df = pd.read_csv(
    filepath_or_buffer='sh000001.csv', # 修改为自己电脑上的csv文件路径
    encoding='gbk',
    parse_dates=['交易日期']
)
print(df)
```

### rolling 计算3天均值

```python
df['三天均值'] = df['收盘价'].rolling(3).mean() # rolling表示只用n行数据进行计算
# 可接max, min, std
print(df)
```

### expanding 计算至今均值

```python
df['至今均值'] = df['收盘价'].expanding().mean()
print(df)
```

### 输出到本地文件

```python
df.to_csv('output.csv', encoding='gbk', index=False) # False不保留index
```
