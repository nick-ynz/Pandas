### 导入数据

```python
import pandas as pd
pd.set_option('expand_frame_repr', False)

df = pd.read_csv(
    filepath_or_buffer='sh000001.csv',
    encoding='gbk',
    # index_col=['交易日期']
    )
print(df)
```

### 对列的操作

```python
print(df['交易日期'].head()+' 15:00:00')
print(df['收盘价'].head()*100)
print((df['收盘价']*df['成交额']).head())
```

```python
df['交易日期2'] = df['交易日期'] + ' 15:00:00'
print(df)
```

### 统计函数

```python
print(df['收盘价'].mean())
print(df[['收盘价','开盘价']].mean())
print(df[['收盘价','开盘价']].mean(1).head()) # 对行求平均
```

```python
print(df['收盘价'].max())
print(df['收盘价'].min())
print(df['收盘价'].std()) # 标准差
print(df['收盘价'].count()) # 非空数据数量
print(df['收盘价'].median()) # 中位数
print(df['收盘价'].quantile(0.25)) # 25%中位数
```

### shift，del函数

```python
df['下周期收盘价'] = df['收盘价'].shift(-1) # 读取上一行数据，并挪动-1格（向上一格），最后留NaN
print(df)
```

```python
del df['下周期收盘价']
del df['交易日期2']
print(df)
```

### diff，drop函数

```python
df['涨跌'] = df['收盘价'].diff(1) # 求本行数据和上一行数据的差
print(df[['收盘价', '涨跌']].head())
```

```python
df.drop(['涨跌'], axis=1, inplace=True) # True的话直接修改df，Flase创建新df再赋值。axis=1代表删除一整列
print(df.head())
```

### pct_change函数

```python
df['涨跌幅'] = df['收盘价'].pct_change(1) # （本行-上n行）/上n行  (2为本行)
print(df[['收盘价', '涨跌幅']].head())
```

### cum类函数

```python
df['累计成交量'] = df['成交量'].cumsum() # 列累计相加
print(df['累计成交量'].head(10))
```

```python
print((df['涨跌幅']+1.0).cumprod().head()) # 累乘
```

### rank函数

```python
df['收盘价排名'] = df['收盘价'].head(15).rank(ascending=True, pct=False) # True从小到大,True前百分之几
print(df[['收盘价', '收盘价排名']].head(15))

df['收盘价排名'] = df['收盘价'].rank(ascending=True, pct=True) # True从小到大,True前百分之几
print('-'*50)
print(df[['收盘价', '收盘价排名']])
```

### counts计数函数

```python
print(df['收盘价'].value_counts().head(10)) # 统计数字出现次数
```
